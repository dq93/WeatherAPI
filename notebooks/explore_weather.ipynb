```python
sns.lineplot(data=merged_data, x='year', y="million_60kgs_bag")
```

The "million_60kgs_bag" variable is increasing indicating that in the later years there was an increase in the amount harvested.

```python
sns.lineplot(data=br_data, x="year", y="million_60kgs_bag", hue="subdivision")
```

Minas Gerais produces far more million_60kgs_bag than the other regions in Brazil between 2002 to 2023.

```python
coffee_totals = br_data.groupby("subdivision")[["million_60kgs_bag", "nonbear_mill_trees"]].sum()
coffee_totals["million_60kgs_bag"].sort_values(ascending=False).plot(kind="bar")
```

The production of Minas Gerais far exceeds the other regions, more than the others shown put together.

```python
sns.lineplot(data=br_data, x="year", y="avg_unemp_perc", hue="subdivision")
```

Only Parana has unemployment info before 2012. When the other regions began recording unemployment, Paranas unemployment percent remained lower than the other regions.

All other regions show a similar trend of rising unemployment around 2013 or 2014 and it peaked around 2016, but then to rose to a max in 2020. Average unemployment percents have fallen since.

```python
weather_vars = ["temperature_2m", "relative_humidity_2m", "precipitation", "surface_pressure"]

for w in weather_vars:
    sns.lineplot(data=merged_data, x="year", y=w)
    plt.ylabel(w.replace('_', ''))
    plt.show()
```

Temperature and surface pressure have risen but relative humidity and precipitation have stayed the same in Minas Gerais within 2022 and 2023.
